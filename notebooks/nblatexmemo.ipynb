## How to enable syntax highlighting for $\LaTeX$ equation in markdown

To enable syntax highlighting in markdown, must enclosed with `$` for inline math, and enclosed with `$$` for displayed math. `\begin{equation}` and `\end{equation}` must be enclosed with `$$` also.

## In "ipynb", just use `\tag` for equation numbering. DO NOT USE `\label` NOR `\eqref`

When numbering equation, just use `\tag`, then refer to the tag MANUALLY.
Why? Because in notebook ("ipynb"), we cannot refer to label in other cell.
(It is possible when using the jupyter notebook, which uses javascript, but now we just solely want to use jupyter lab!)

For inline math, enclosed the with `$` ... `$`.
For displayed math, enclose with `$$` ... `$$`.

Enclose `\begin{align}`, `\begin{equation}` with `$$` in order to activate syntax highlighting.

In notebook markdown cell, do NOT use "tab" or insert space after `\begin{equation}`.

## vscode is using katex, jupyterlab is using MathJax as $\LaTeX$ renderer

Install vscode extension: Markdown PDF, markdownlint (BAD, this is just for markdown file, NOT notebook ipynb file).
This is to make the view in jupyterlab (which uses MathJax) and vscode (which uses katex) to be the same. BUT NOT POSSIBLE!

So, use $\LaTeX$ macro that can be rendered by both katex and MathJax!
